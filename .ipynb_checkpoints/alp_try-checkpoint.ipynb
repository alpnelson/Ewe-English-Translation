```python
english_data = []
with open('jw300.en-ee.en',encoding="utf8") as file:

    line = file.readline()
    cnt = 1
    while line:
        english_data.append(line.strip())
        line = file.readline()
        cnt += 1


twi_data = []
with open('jw300.en-ee.ee',encoding="utf8") as file:

    # twi=file.read()
    line = file.readline()
    cnt = 1
    while line:
        twi_data.append(line.strip())
        line = file.readline()
        cnt += 1
```

```python
import pandas as pd
df = pd.DataFrame({'Ewe': twi_data, 'English': english_data})
df.head(20)
```
